```python
#  [Problem 1] Data acquisition
from sklearn.datasets import load_iris
import pandas as pd
iris = load_iris()
# X is the explanatory variable
# Y is the objective variable
Xt, Yt = iris.data, iris.target 
X = pd.DataFrame(Xt, columns = ['sepal_length','sepal_width','petal_length','petal_width'])
Y = pd.DataFrame(Yt, columns = ['Species'])
print('-----------------------------------Problem #1-------------------------------------')
print(X.shape)
print(Y.shape)
print(X)
print(Y)
#  [Problem 2] Combining data 
print('-----------------------------------Problem #2-------------------------------------')
frames = [X, Y]
df = pd.concat(frames, axis = 1)
print(df)
# [Problem 3] Checking the data 
print('-----------------------------------Problem #3.1-------------------------------------')
# Display the 4th sample from the beginning and see what each feature is like. (Numerical data or character data, etc.)
print(df.loc[4,:])
print('-----------------------------------Problem #3.2-----------------------------------')
# Output the total number of samples for each label.
print(df.info(verbose=True))
print('-----------------------------------Problem #3.3-----------------------------------')
# Check if there is a missing value in the feature quantity.
print(df.isnull())
print('-----------------------------------Problem #3.4-----------------------------------')
# Display the mean, standard deviation, and quartiles of the feature values at once.
df0 = df.loc[df.Species == 0] # Iris setosa
print(df0[['sepal_length','sepal_width','petal_length','petal_width']].describe())

df1 = df.loc[df.Species == 1] # Iris virgiclor
print(df1[['sepal_length','sepal_width','petal_length','petal_width']].describe())

df2 = df.loc[df.Species == 2] # Iris virginica
print(df2[['sepal_length','sepal_width','petal_length','petal_width']].describe())

# [Problem 4] Examining the iris dataset itself
# The mean value is the highes for Iris virginica and lowest for Iris setosa
# [Problem 5] Extracting the required data
print('-----------------------------------Problem #5.1-----------------------------------')
# Extract the sepal_width column in two different ways
print(df['sepal_width'])
print(df.iloc[:,1])

print('-----------------------------------Problem #5.2-----------------------------------')
# Extract the 50th to 99th data
print(df.loc[50:99,:])

print('-----------------------------------Problem #5.3-----------------------------------')
# Extract the50th to 99th data of the petal_length column
print(df.loc[50:99,'petal_length'])

print('-----------------------------------Problem #5.4-----------------------------------')
# Extract data with apetal_width value of 0.2
print(df.loc[df.petal_length == 0.2])

# [Problem 6] Creating a diagram
print('-----------------------------------Problem #6.1-----------------------------------')
# Make a pie chart of the number of samples per label (while showing percentages)
import matplotlib.pyplot as plt
import numpy as np

NSL = df[['petal_length']].size
NSW = df[['sepal_width']].size
NPL = df[['petal_length']].size
NPW = df[['petal_width']].size

irislabels = ['sepal_length','sepal_width','petal_length','petal_width']
plt.pie([NSL, NSW, NPL, NPW], labels = irislabels, autopct='%1.1f%%')
plt.show() 

print('-----------------------------------Problem #6.2-----------------------------------')
# Select features one by one and visualize the distribution of data for each label using a box plot
# the same goes for the standard deviation (std)


plt.boxplot(df0[['sepal_length','sepal_width','petal_length','petal_width']], labels = irislabels)
plt.title('Iris setosa')
plt.xlabel('xlabel')
plt.ylabel('ylabel')
plt.show() 

plt.boxplot(df1[['sepal_length','sepal_width','petal_length','petal_width']], labels = irislabels)
plt.title('Iris virgiclor')
plt.xlabel('xlabel')
plt.ylabel('ylabel')
plt.show() 

plt.boxplot(df2[['sepal_length','sepal_width','petal_length','petal_width']], labels = irislabels)
plt.title('Iris virginica')
plt.xlabel('xlabel')
plt.ylabel('ylabel')
plt.show() 
     
    
print('-----------------------------------Problem #6.3-----------------------------------')
# Select features one by one and visualize the distribution of data for each label using a violin plot
plt.figure()
ax = plt.subplot(111)
pos   = [1, 2, 3, 4]
plt.violinplot(df0[['sepal_length','sepal_width','petal_length','petal_width']],pos, vert=False)
plt.title('Iris setosa')
plt.xlabel('xlabel')
plt.ylabel('ylabel')
ax.set_yticks(pos)
ax.set_yticklabels(irislabels)
plt.show() 

plt.violinplot(df1[['sepal_length','sepal_width','petal_length','petal_width']],pos, vert=False)
plt.title('Iris virgiclor')
plt.xlabel('xlabel')
plt.ylabel('ylabel')
ax.set_yticks(pos)
ax.set_yticklabels(irislabels)
plt.show() 

plt.violinplot(df2[['sepal_length','sepal_width','petal_length','petal_width']],pos, vert=False)
plt.title('Iris virginica')
plt.xlabel('xlabel')
plt.ylabel('ylabel')
plt.show() 

# [Problem 7] Confirming the relationship between features
print('-----------------------------------Problem #7.1-----------------------------------')
# Make scatter plots of all combinations (six), such as choosing two from four features, color-coded by type.
# The following figure is an example of one of them.
plt.scatter(df0[['sepal_length']],df0[['sepal_width']])
plt.scatter(df1[['sepal_length']],df1[['sepal_width']])
plt.scatter(df2[['sepal_length']],df2[['sepal_width']])
plt.xlabel('sepal_length')
plt.ylabel('sepal_width')
plt.legend(['setosa' , 'virgiclor', 'virginica'])
plt.show() 

plt.scatter(df0[['petal_length']],df0[['petal_width']])
plt.scatter(df1[['petal_length']],df1[['petal_width']])
plt.scatter(df2[['petal_length']],df2[['petal_width']])
plt.xlabel('petal_length')
plt.ylabel('petal_width')
plt.legend(['setosa' , 'virgiclor', 'virginica'])
plt.show() 

print('-----------------------------------Problem #7.2-----------------------------------')
# Create a scatterplot matrix that displays all combinations of features at once.
import seaborn as sns
sns.set(style="ticks", color_codes=True)
g = sns.pairplot(df0[['sepal_length','sepal_width','petal_length','petal_width']])
plt.title('Iris setosa')
plt.show()

g = sns.pairplot(df1[['sepal_length','sepal_width','petal_length','petal_width']])
plt.title('Iris virgiclor')
plt.show()

g = sns.pairplot(df2[['sepal_length','sepal_width','petal_length','petal_width']])
plt.title('Iris virginica')
plt.show()

print('-----------------------------------Problem #7.3-----------------------------------')
# Create a correlation coefficient matrix for 4 features The correlation coefficient matrix can be obtained by using corr()
df0.corr(method='pearson')
df1.corr(method='pearson')
df2.corr(method='pearson')

print('-----------------------------------Problem #7.4-----------------------------------')
# Make a heat map of the correlation coefficient matrix (1)
sns.heatmap(df0.corr())
plt.show()

sns.heatmap(df1.corr())
plt.show()

sns.heatmap(df2.corr())
plt.show()


# [Problem 8] Explaining the results of visualized graphs and tables
print('-----------------------------------Problem #8.0-----------------------------------')
# for the case of Iris setosa
# There is strong correlation between sepal width and sepal length as well as between sepal length and sepal width
# and slight correlation between petal width and petal length as well as between petal length and petal width

# for the case of Iris virgiclor
# there is storng correlation between petal width and sepal length as well as between petal length and petal width
# and slight correlation between petal length and sepal length as well as between sepal length and petal length
 
# for the case of Iris virginica
# There is strong correlation between petal length and sepal length as wll as between sepal length and petal length
# and  slight correlation between petal width and sepal width as well as between sepal width and petal width
```
